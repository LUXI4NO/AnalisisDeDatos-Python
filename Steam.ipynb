# ¿Cuál es el promedio de horas de uso para todos los juegos?


```python
# Calculo el promedio de horas por usuario
promedio_horas = df['avg_hours_per_user'].mean()
print(f'Promedio de horas por usuario: {promedio_horas:.2f} horas')
```

```python
# Creo una figura con un tamaño específico y una resolución determinada
figura = plt.figure(figsize=(12, 6.75), dpi=90, facecolor="#ECECEC")

# Añado un eje al gráfico con un color de fondo específico
eje = plt.subplot(111, facecolor="#ECECEC")

# Oculto los bordes superior y derecho del gráfico
eje.spines["top"].set_visible(False)
eje.spines["right"].set_visible(False)

# Añado una rejilla al gráfico con un estilo específico y un color
eje.grid(True, color="lightgrey", linestyle=":")

# Creo el gráfico de barras en el eje
barras = eje.bar('Promedio de Horas por Usuario', 
                 promedio_horas, 
                 edgecolor='black', 
                 linewidth=0.75, 
                 color='#2a475e', 
                 zorder=3, 
                 width=0.75)

# Añado título y etiquetas con formato mejorado
eje.set_title('Promedio de Horas por Usuario en Juegos de Steam', fontsize=16, weight='bold')
eje.set_ylabel('Horas', fontsize=12, weight='bold')

# Anoto cada barra con su valor correspondiente
for barra in barras:
    altura = barra.get_height()
    eje.annotate(f'{altura:.2f} hrs', 
                 xy=(barra.get_x() + barra.get_width() / 2, altura), 
                 xytext=(0, 5),  
                 textcoords="offset points",
                 ha='center', va='bottom',
                 fontsize=8, weight='bold', color='#333333')

# Muestro el gráfico
plt.show()
```

# ¿Cuáles son los géneros de videojuegos más vendidos?

```python
# Luego, sumo las ventas por género
generos = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing', 'free_to_play', 'early_access']
ventas_por_genero = {genero: 0 for genero in generos}

for genero in generos:
    ventas_por_genero[genero] = df[df[genero]]['sells'].sum()

# Ordeno los resultados
ventas_ordenadas = sorted(ventas_por_genero.items(), key=lambda x: x[1], reverse=True)

# Muestro los resultados
print("Géneros más vendidos:")
for genero, ventas in ventas_ordenadas:
    print(f"{genero.capitalize()}: {ventas:,.0f} ventas")

# Preparo los datos para el gráfico
generos_capitalizados = [item[0].capitalize() for item in ventas_ordenadas]
ventas = [item[1] for item in ventas_ordenadas]
```

```python
# Creo una figura con un tamaño específico y una resolución determinada
figura = plt.figure(figsize=(12, 6.75), dpi=90, facecolor="#ECECEC")

# Añado un eje al gráfico con un color de fondo específico
eje = plt.subplot(111, facecolor="#ECECEC")

# Oculto los bordes superior y derecho del gráfico
eje.spines["top"].set_visible(False)
eje.spines["right"].set_visible(False)

# Añado una rejilla al gráfico con un estilo específico y un color
eje.grid(True, color="lightgrey", linestyle=":")

# Creo el gráfico de barras
barras = eje.bar(generos_capitalizados,
                 ventas,
                 edgecolor='Black',
                 linewidth=0.75,
                 color='#2a475e',
                 zorder=3,
                 width=0.75)

# Añado título y etiquetas con formato mejorado
plt.title('Ventas Totales por Género', fontsize=16, weight='bold', pad=20)
eje.set_xlabel('Género', fontsize=12, weight='bold')
eje.set_ylabel('Ventas Totales', fontsize=12, weight='bold')

# Añado números sobre las barras
for barra in barras:
    altura = barra.get_height()
    eje.annotate(f'{altura:,.0f}', 
                 xy=(barra.get_x() + barra.get_width() / 2, altura), 
                 xytext=(0, 3), 
                 textcoords="offset points",
                 ha='center', va='bottom',
                 fontsize=8, weight='bold', color='#333333')

# Establezco la escala y el formato de los ejes
eje.yaxis.set_major_locator(plt.MultipleLocator(250000000))
eje.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:,.0f}"))

# Ajusto el diseño para que todo se muestre correctamente
plt.tight_layout()

# Finalmente, muestro el gráfico
plt.show()
```

# La relación entre precio y ventas de desarrolladores

```python
# Primero, sumo las ventas por desarrollador
ventas_por_desarrollador = df.groupby('developer')['sells'].sum()

# Obtengo los 10 desarrolladores con más ventas
top_10_desarrolladores = ventas_por_desarrollador.nlargest(10)

# Filtrado del DataFrame para obtener solo los juegos de los 10 desarrolladores principales
df_top_10 = df[df['developer'].isin(top_10_desarrolladores.index)]

# Muestro los resultados
print("Top 10 desarrolladores con más ventas:")
for developer, total_ventas in top_10_desarrolladores.items():
    print(f"{developer}: {total_ventas:,.0f} ventas")
```

```python
# Creo una figura con un tamaño específico y una resolución determinada
figura = plt.figure(figsize=(12, 6.75), dpi=90, facecolor="#ECECEC")

# Añado un eje al gráfico con un color de fondo específico
eje = plt.subplot(111, facecolor="#ECECEC")

# Oculto los bordes superior y derecho del gráfico
eje.spines["top"].set_visible(False)
eje.spines["right"].set_visible(False)

# Añado una rejilla al gráfico con un estilo específico y un color
eje.grid(True, color="lightgrey", linestyle=":")

# Creo el gráfico de dispersión
dispersion = eje.scatter(df_top_10['price'], df_top_10['sells'], color='#66c0f4', edgecolor='black', s=100)

# Añado etiquetas y título
eje.set_xlabel('Precio (€)', fontsize=12, weight='bold')
eje.set_ylabel('Total de Ventas', fontsize=12, weight='bold')
eje.set_title('Relación entre Precio y Ventas de los 10 Desarrolladores con Más Ventas', fontsize=14, weight='bold', pad=20)

# Establezco la escala y el formato de los ejes
eje.yaxis.set_major_locator(MultipleLocator(20000000))
eje.xaxis.set_major_locator(MultipleLocator(10))
eje.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{int(y):,}"))
eje.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:,.0f}€"))

# Añado nombres de los 3 desarrolladores con más ventas
etiquetas = []
for i, desarrollador in enumerate(top_10_desarrolladores.index[:3]):
    # Filtro los datos correspondientes al desarrollador
    df_dev = df_top_10[df_top_10['developer'] == desarrollador]
    
    # Añado etiquetas de texto debajo de los puntos
    for _, fila in df_dev.iterrows():
        etiqueta = eje.text(fila['price'], fila['sells'], desarrollador, 
                            ha='center', va='top', fontsize=8, color='black')
        etiquetas.append(etiqueta)

# Ajusto la posición de las etiquetas para evitar superposiciones
adjust_text(etiquetas, only_move={'points': 'y', 'text': 'y'}, arrowprops=dict(arrowstyle='-', color='gray'))

# Ajusto el diseño para que todo se muestre correctamente
plt.tight_layout()

# Finalmente, muestro el gráfico
plt.show()
```

# ¿Qué género tiene la mayor cantidad de valoraciones positivas y cuál tiene la mayor cantidad de valoraciones negativas?

```python
# Sumar las valoraciones por género
generos = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing', 'free_to_play', 'early_access']

valoraciones_positivas = {genero: 0 for genero in generos}
valoraciones_negativas = {genero: 0 for genero in generos}

for genero in generos:
    valoraciones_positivas[genero] = df[df[genero]]['positive_ratings'].sum()
    valoraciones_negativas[genero] = df[df[genero]]['negative_ratings'].sum()

# Ordenar los resultados
valoraciones_ordenadas_positivas = sorted(valoraciones_positivas.items(), key=lambda x: x[1], reverse=True)
valoraciones_ordenadas_negativas = sorted(valoraciones_negativas.items(), key=lambda x: x[1], reverse=True)

# Mostrar los resultados
print("Géneros con más valoraciones positivas:")
for genero, valoracion_positiva in valoraciones_ordenadas_positivas:
    print(f"{genero.capitalize()}: {valoracion_positiva:,} valoraciones positivas")

print("\nGéneros con más valoraciones negativas:")
for genero, valoracion_negativa in valoraciones_ordenadas_negativas:
    print(f"{genero.capitalize()}: {valoracion_negativa:,} valoraciones negativas")

# Valores de las valoraciones por género
valores_positivos = list(valoraciones_positivas.values())
valores_negativos = list(valoraciones_negativas.values())
generos = list(valoraciones_positivas.keys())
```

```python
# Creo una figura con un tamaño específico y una resolución determinada
figura = plt.figure(figsize=(12, 6.75), dpi=90, facecolor="#ECECEC")

# Añado un eje al gráfico con un color de fondo específico
eje = plt.subplot(111, facecolor="#ECECEC")

# Oculto los bordes superior y derecho del gráfico
eje.spines["top"].set_visible(False)
eje.spines["right"].set_visible(False)

# Añado una rejilla al gráfico con un estilo específico y un color
eje.grid(True, color="lightgrey", linestyle=":")

# Gráfico de barras para valoraciones positivas
barras_positivas = plt.bar(generos, valores_positivos, color='#66c0f4', ec='Black', lw=0.75, zorder=3, alpha=0.7, width=0.75, label='Valoraciones Positivas')

# Gráfico de barras para valoraciones negativas
barras_negativas = plt.bar(generos, valores_negativos, color='#2a475e', ec='Black', lw=0.75, zorder=3, alpha=0.7, width=0.75, label='Valoraciones Negativas')

# Añadir etiquetas y título
# Añado etiquetas y título
eje.set_xlabel('Géneros', fontsize=12, weight='bold')
eje.set_ylabel('Cantidad de Valoraciones', fontsize=12, weight='bold')
eje.set_title('Valoraciones Positivas y Negativas por Género', fontsize=14, weight='bold', pad=20)
eje.legend()

# Establecer la escala y el formato de los ejes
eje.yaxis.set_major_locator(plt.MultipleLocator(2000000))
eje.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:,.0f}"))

# Mostrar el valor de cada barra encima de la misma
for barra in barras_positivas:
    altura = barra.get_height()
    eje.text(barra.get_x() + barra.get_width() / 2.0, altura, f'{altura:,.0f}', ha='center', va='bottom', fontsize=8, color='black', weight='bold')

for barra in barras_negativas:
    altura = barra.get_height()
    eje.text(barra.get_x() + barra.get_width() / 2.0, altura, f'{altura:,.0f}', ha='center', va='bottom', fontsize=8, color='black', weight='bold')

# Mostrar gráfico
plt.show()
```

# ¿Cuáles son las horas jugadas mensualmente por tipo de juego?

```python
# Creo una nueva columna para calcular las horas totales jugadas por juego
df['horas_totales'] = df['avg_hours_per_user'] * df['sells']

# Defino una lista con los tipos de juegos
tipos_de_juegos  = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing', 'free_to_play', 'early_access']

# Multiplico el total de horas jugadas por cada tipo de juego
for tipo_de_juego  in tipos_de_juegos :
    df[tipo_de_juego  + '_horas'] = df[tipo_de_juego ] * df['horas_totales']

# Agrupo por mes y sumo las horas jugadas por cada tipo de juego
horas_por_mes  = df.groupby('month')[[tipo_de_juego  + '_horas' for tipo_de_juego  in tipos_de_juegos ]].mean()

# Convierto el índice numérico de meses a nombres de meses
horas_por_mes.index = horas_por_mes.index.map({
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
})

# Creo una copia del DataFrame para aplicar el formato sin modificar los datos originales
horas_por_mes_formateado = horas_por_mes.copy()

# Aplico el formato .0f a todos los valores en el DataFrame usando map()
horas_por_mes_formateado = horas_por_mes_formateado.apply(lambda x: x.map(lambda y: f"{y:,.0f}"))

# Muestro el DataFrame resultante con las horas jugadas por tipo de juego, agrupadas por mes
print("Horas jugadas por tipo de juego, agrupadas por mes:")
print(horas_por_mes_formateado)
```

```python
# Creo una figura con un tamaño específico y una resolución determinada
figura = plt.figure(figsize=(12, 6.75), dpi=90, facecolor="#ECECEC")

# Añado un eje al gráfico con un color de fondo específico
eje = plt.subplot(111, facecolor="#ECECEC")

# Oculto los bordes superior y derecho del gráfico
eje.spines["top"].set_visible(False)
eje.spines["right"].set_visible(False)

# Añado una rejilla al gráfico con un estilo específico y un color
eje.grid(True, color="lightgrey", linestyle=":")

# Creo un gráfico de áreas apiladas usando los datos agrupados por mes
horas_por_mes.plot(kind='area', stacked=True, ax=eje, color=plt.cm.Paired.colors)

# Añado etiquetas y título al gráfico
eje.set_xlabel('Mes', fontsize=12, weight='bold')
eje.set_ylabel('Total de Horas Jugadas', fontsize=12, weight='bold')
eje.set_title('Horas Jugadas Mensualmente por Tipo de Juego', fontsize=14, weight='bold')

# Roto las etiquetas del eje X para mejorar la legibilidad
plt.xticks(rotation=0, fontsize=10)

# Añado una leyenda fuera del gráfico para una mejor visualización
eje.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10, frameon=False)

# Establezco la escala y el formato de los ejes
eje.yaxis.set_major_locator(plt.MultipleLocator(1000000000))
eje.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:,.0f}"))

# Ajusto el diseño para que todos los elementos del gráfico se muestren correctamente
plt.tight_layout()

# Muestro el gráfico
plt.show()
```
